```python
#dictionary of desired values
houses = {'city': [],
         'zipcode': [],
         'state': [],
         'county': [],
         'longitude': [],
         'latitude': [],
         'sqft': [],
        }

#populate dict values and append to list
for house in json_data:
    houses['city'].append(house['location']['address']['city'])
    houses['zipcode'].append(house['location']['address']['postal_code'])
    houses['state'].append(house['location']['address']['state'])
    houses['county'].append(house['location']['county']['name'])
    houses['longitude'].append(house['location']['address']['coordinate']['lon'])
    houses['latitude'].append(house['location']['address']['coordinate']['lat'])
    houses['sqft'].append(house['description']['sqft'])
    #house_list.append(houses)
    


print(houses)
```
